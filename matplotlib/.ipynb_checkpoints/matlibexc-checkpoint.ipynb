```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y) #(1,1),(2,4),(3,9)...
plt.show()
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y) #(1,1),(2,4),(3,9)...
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y,marker='*') #(1,1),(2,4),(3,9)...
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y,marker='o',linestyle='--') #(1,1),(2,4),(3,9)...
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```
