```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,50)
y = x**2
plt.plot(x,y) #(1,1),(2,4),(3,9)...
plt.show()
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y) #(1,1),(2,4),(3,9)...
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y,marker='o') #(1,1),(2,4),(3,9)...
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y,marker='o',linestyle='--') #(1,1),(2,4),(3,9)...
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y,marker='o',linestyle='-',color='#0000FF') #(1,1),(2,4),(3,9)...
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()

```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y,'o-r') #(1,1),(2,4),(3,9)...
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```

```python
x = np.arange(1,6)
plt.plot(x,x**3,'o-r',alpha=0.7)
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```

```python
# linewidth and marker size can be represented by using 'lw' and 'ms'
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,6)
plt.plot(x,x**3,'o-g',lw=10,ms=15) # default blue color
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```

```python
x = np.arange(1,6)
plt.plot(x,x**3,'o-g',lw=10,ms=15,mfc='yellow') # default blue color
plt.title('Square Function Line Plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```
